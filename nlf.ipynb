## Visualizing Passing Yards by Team

```python
passing_by_team = weekly.reset_index().merge(team, left_on='recent_team', right_on='team_abbr', how='left')
passing_yards_color = passing_by_team[passing_by_team['week'].between(1, 12)].groupby('recent_team')['passing_yards'].sum()
passing_yards_team = passing_yards_color.sort_values(ascending=False)
team_color_dict = dict(zip(team['team_abbr'], team['team_color']))
fig, ax = plt.subplots(figsize=(12, 8))
bars = passing_yards_team.plot(kind='bar', color=[team_color_dict.get(team, '#000000') for team in passing_yards_team.index], ax=ax)
plt.title('Total Passing Yards by Team (Weeks 1-12)')
plt.xlabel('Team')
plt.ylabel('Total Passing Yards')
plt.show(bars)
```

### Explosive Play Analysis for Giants

```python
giants_pbp = pbp[pbp['game_id'].str.contains('NYG')]
giants_pbp_clean = giants_pbp[['play_id','game_id','home_team','away_team','week',
                        'posteam','posteam_type','defteam','side_of_field', 'yardline_100',
                        'drive', 'sp', 'qtr', 'down', 'goal_to_go', 'time', 
                        'yrdln', 'ydstogo','play_type', 'yards_gained','offense_formation', 
                        'offense_personnel', 'defenders_in_box', 'defense_personnel', 'number_of_pass_rushers']]
giants_field = giants_pbp_clean[giants_pbp_clean['play_type'].notna()]
giants_offense = giants_field[giants_field['posteam'] == 'NYG']
```

```python
def is_explosive(row):
    if (row['yards_gained'] >= 20 and row['play_type'] == 'pass'):
        return True
    elif (row['yards_gained'] >= 10 and row['play_type'] == 'run'):
        return True
    else:
        return False
    
giants_offense['explosive_play'] = giants_offense.apply(is_explosive, axis=1)
```

```python
nyg_explosive = giants_offense[giants_offense['explosive_play'] ==  True]
sns.countplot(data=nyg_explosive, x='offense_formation', hue='explosive_play')
plt.title('Explosive Play by Formation')
plt.xlabel('Formation')
plt.ylabel('Count')
plt.show()
```
